```python
import pandas as pd
import numpy as np
pd.options.display.max_rows=125
```

```python
df=pd.read_csv('Minneapolis Open Data/Weather Data.csv')
```

```python
df.columns
```

```python
df=df[['DATE', 'DailyMaximumDryBulbTemperature','DailyAverageDryBulbTemperature']]
```

```python
df=df[pd.notnull(df.DailyMaximumDryBulbTemperature)]
```

```python
df.DATE.nunique()
```

```python
df.dtypes
```

```python
df['DATE']=pd.to_datetime(df.DATE)
```

```python
df.groupby([df.DATE.dt.month,df.DATE.dt.year]).aggregate('mean')
```

```python
df=df.set_index('DATE')
```

```python
df['threeDayAvgMaxWeather']=df.DailyMaximumDryBulbTemperature.rolling(3).mean()
```

```python
df['weatherCalcDate']=df.index
```

```python
df=df.shift(1).reset_index()
```

```python
df=df.rename(columns={'DATE':'inspectionDate'})
```

```python
df['inspectionDate']=pd.to_datetime(df.inspectionDate).dt.date
```

```python
df=df[['inspectionDate','threeDayAvgMaxWeather']]
```
